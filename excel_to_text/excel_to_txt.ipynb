```python
import pandas as pd

# Specify the path to the Excel file
input_excel_path = 'input.xlsx'  # Change this to your actual Excel file path
output_txt_path = 'output.txt'  # Path to save the output text file

# Read the Excel file
try:
    df = pd.read_excel(input_excel_path)
except FileNotFoundError:
    print("The specified Excel file was not found.")
    exit()
except Exception as e:
    print(f"An error occurred while reading the Excel file: {e}")
    exit()

# Write the content to a text file
with open(output_txt_path, 'w', encoding='utf-8') as file:
    # Convert the DataFrame to a string and write it to the text file
    file.write(df.to_string(index=False))

print(f"Excel content has been successfully written to {output_txt_path}")
```
