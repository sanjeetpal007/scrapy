```python
import pdfplumber
from docx import Document
#pip install pydocx
def convert_pdf_to_word(pdf_path, word_path):
    # Create a new Word document
    doc = Document()
    
    # Open the PDF file using pdfplumber
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            # Extract the text from each page
            text = page.extract_text()
            if text:
                # Add the extracted text as a new paragraph to the Word document
                doc.add_paragraph(text)
    
    # Save the Word document to the specified path
    doc.save(word_path)
    print(f"PDF has been successfully converted to Word: {word_path}")

# Replace 'input.pdf' and 'output.docx' with your file paths
input_pdf_path = 'page_no_change.pdf'
output_word_path = 'output.docx'

# Call the function to convert PDF to Word
convert_pdf_to_word(input_pdf_path, output_word_path)
```

```python
import pytesseract
from pdf2image import convert_from_path
from docx import Document
from PIL import Image
import os

# Specify the path to the Tesseract OCR executable if not in your PATH
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

def convert_pdf_images_to_word(pdf_path, word_path):
    try:
        # Convert PDF to a list of images (one per page)
        pages = convert_from_path(pdf_path, 300)  # The 300 dpi setting for better OCR accuracy
        
        # Create a new Word document
        doc = Document()
        
        # Process each image (each page of the PDF)
        for i, page in enumerate(pages):
            # Use pytesseract to extract text from the image
            text = pytesseract.image_to_string(page)
            
            # Add the extracted text to the Word document
            if text.strip():  # Check if there's any text extracted
                doc.add_paragraph(text)
            
            # Optionally, you can add a page break after each page's content
            if i < len(pages) - 1:
                doc.add_page_break()
        
        # Save the Word document
        doc.save(word_path)
        print(f"PDF with images has been successfully converted to Word: {word_path}")
    except FileNotFoundError as e:
        print(f"File error: {e}")
    except pytesseract.TesseractNotFoundError:
        print("Tesseract OCR is not found. Please make sure Tesseract is installed and the path is set correctly.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Replace with your file paths
input_pdf_path = 'tester.pdf'
output_word_path = 'output.docx'

# Convert the PDF to a Word document
convert_pdf_images_to_word(input_pdf_path, output_word_path)
```

```python
import os
from pdf2image import convert_from_path
from pdfminer.high_level import extract_text
from docx import Document
from docx.shared import Inches
from PIL import Image
import re

# Function to extract text from a PDF file
def pdf_to_text(pdf_file):
    text = extract_text(pdf_file)
    return text

# Function to clean text to make it XML compatible
def clean_text(text):
    if text is None:
        return ""
    # Remove any non-printable characters and null bytes
    return re.sub(r'[\x00-\x1F\x7F]', '', text)

# Function to extract images from a PDF and save them as image files
def pdf_to_images(pdf_file, image_folder="extracted_images"):
    images = convert_from_path(pdf_file)
    if not os.path.exists(image_folder):
        os.makedirs(image_folder)
    image_paths = []
    for i, image in enumerate(images):
        image_path = os.path.join(image_folder, f"page_{i + 1}.png")
        image.save(image_path, "PNG")
        image_paths.append(image_path)
    return image_paths

# Function to save text and images to a Word document
def create_word_document(text, image_paths, output_file):
    document = Document()
    
    # Add the extracted text
    if text:
        document.add_paragraph(text)
    
    # Add images to the Word document
    for image_path in image_paths:
        document.add_picture(image_path, width=Inches(6))
    
    document.save(output_file)
    print(f"Word document saved as {output_file}")

# Main function to convert PDF to Word with images
def convert_pdf_to_word_with_images(input_pdf, output_word):
    print(f"Converting {input_pdf} to {output_word}...")
    text = pdf_to_text(input_pdf)
    cleaned_text = clean_text(text)
    image_paths = pdf_to_images(input_pdf)
    create_word_document(cleaned_text, image_paths, output_word)
    print("Conversion completed.")

# Example usage
if __name__ == "__main__":
    input_pdf = 'page_no_change.pdf'  # Replace with your PDF file path
    output_word = 'output.docx'  # Replace with your desired output file name
    convert_pdf_to_word_with_images(input_pdf, output_word)
```

```python
import os
import fitz  # PyMuPDF
from pdfminer.high_level import extract_text
from docx import Document
from docx.shared import Inches
from PIL import Image
import re

# Function to clean text and remove non-XML compatible characters
def clean_text(text):
    if text is not None:
        text = re.sub(r'[\x00-\x1F\x7F]', '', text)  # Remove control characters
        text = text.replace('\x0b', '').replace('\x0c', '')  # Remove specific problematic characters
    return text

# Function to extract text from a PDF using pdfminer
def extract_text_from_pdf(pdf_file):
    text = extract_text(pdf_file)
    return clean_text(text)

# Function to extract images from a PDF and save them as image files
def extract_images_from_pdf(pdf_file, image_folder="extracted_images"):
    if not os.path.exists(image_folder):
        os.makedirs(image_folder)

    document = fitz.open(pdf_file)
    image_paths = []

    for page_num in range(len(document)):
        page = document.load_page(page_num)
        images = page.get_images(full=True)

        for img_index, image_info in enumerate(images):
            xref = image_info[0]  # The first value is the xref of the image
            image = document.extract_image(xref)
            image_bytes = image["image"]
            image_ext = image["ext"]
            image_path = os.path.join(image_folder, f"page_{page_num + 1}_img_{img_index + 1}.{image_ext}")

            with open(image_path, "wb") as img_file:
                img_file.write(image_bytes)

            image_paths.append(image_path)

    return image_paths

# Function to add text and images to a Word document
def create_word_document(text, image_paths, output_file):
    document = Document()

    # Add the extracted text to the document
    if text:
        document.add_paragraph(text)

    # Add images to the Word document
    for image_path in image_paths:
        try:
            document.add_picture(image_path, width=Inches(6))
        except Exception as e:
            print(f"Failed to add image {image_path}: {e}")

    document.save(output_file)
    print(f"Word document saved as {output_file}")

# Main function to convert PDF to Word with images and text
def convert_pdf_to_word(input_pdf, output_word):
    print(f"Converting {input_pdf} to {output_word}...")
    try:
        text = extract_text_from_pdf(input_pdf)
        image_paths = extract_images_from_pdf(input_pdf)
        create_word_document(text, image_paths, output_word)
        print("Conversion completed.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage
if __name__ == "__main__":
    input_pdf = 'my_file_with_footer4.pdf'  # Replace with your PDF file path
    output_word = 'output.docx'  # Replace with your desired output file name
    convert_pdf_to_word(input_pdf, output_word)
```

```python
import os
import fitz  # PyMuPDF
from pdfminer.high_level import extract_text
from docx import Document
from docx.shared import Inches
from PIL import Image
import re

# Function to clean text and remove non-XML compatible characters
def clean_text(text):
    if text is not None:
        # Remove control characters and non-XML compatible characters
        text = re.sub(r'[\x00-\x1F\x7F]', '', text)  # Remove control characters
        text = text.replace('\x0b', '').replace('\x0c', '')  # Remove specific problematic characters
    return text

# Function to extract text from a PDF using pdfminer
def extract_text_from_pdf(pdf_file):
    text = extract_text(pdf_file)
    return clean_text(text)

# Function to extract images from a PDF and save them as image files
def extract_images_from_pdf(pdf_file, image_folder="extracted_images"):
    if not os.path.exists(image_folder):
        os.makedirs(image_folder)

    document = fitz.open(pdf_file)
    image_paths = []

    for page_num in range(len(document)):
        page = document.load_page(page_num)
        images = page.get_images(full=True)

        for img_index, image_info in enumerate(images):
            xref = image_info[0]  # The first value is the xref of the image
            image = document.extract_image(xref)
            image_bytes = image["image"]
            image_ext = image["ext"]
            image_path = os.path.join(image_folder, f"page_{page_num + 1}_img_{img_index + 1}.{image_ext}")

            with open(image_path, "wb") as img_file:
                img_file.write(image_bytes)

            image_paths.append(image_path)

    return image_paths

# Function to add text and images to a Word document
def create_word_document(text, image_paths, output_file):
    document = Document()

    # Add the extracted text to the document
    if text:
        document.add_paragraph(text)

    # Add images to the Word document
    for image_path in image_paths:
        try:
            document.add_picture(image_path, width=Inches(6))
        except Exception as e:
            print(f"Failed to add image {image_path}: {e}")

    document.save(output_file)
    print(f"Word document saved as {output_file}")

# Main function to convert PDF to Word with images and text
def convert_pdf_to_word(input_pdf, output_word):
    print(f"Converting {input_pdf} to {output_word}...")
    try:
        # Extract and clean text from PDF
        text = extract_text_from_pdf(input_pdf)
        if not text:
            print("No text found in the PDF.")
        
        # Extract images from PDF
        image_paths = extract_images_from_pdf(input_pdf)
        if not image_paths:
            print("No images found in the PDF.")

        # Create Word document
        create_word_document(text, image_paths, output_word)
        print("Conversion completed.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage
if __name__ == "__main__":
    input_pdf = "C:\\Users\\hp.DESKTOP-MTCIJR8\\Videos\\dfccil\\document.pdf"  # Replace with your PDF file path
    output_word = 'output.docx'  # Replace with your desired output file name
    convert_pdf_to_word(input_pdf, output_word)
```

```python
from pdf2docx import Converter

def convert_pdf_to_word(input_pdf, output_word):
    # Create a PDF to Word converter object
    try:
        converter = Converter(input_pdf)
        # Convert the entire PDF to a Word document
        converter.convert(output_word, start=0, end=None)
        converter.close()
        print(f"PDF file '{input_pdf}' has been successfully converted to Word file '{output_word}'.")
    except Exception as e:
        print(f"An error occurred during the conversion: {e}")

# Example usage
if __name__ == "__main__":
    input_pdf = 'C:\\Users\\hp.DESKTOP-MTCIJR8\\Videos\\dfccil\\document.pdf'  # Replace with your PDF file path
    output_word = 'output.docx'  # Replace with your desired output file name
    convert_pdf_to_word(input_pdf, output_word)
```
